```python
import yfinance as yf
import numpy as np
import pandas as pd

ticker = "NBIS"

prices = yf.download(
    ticker,
    period="3y",
    interval="1d",
    auto_adjust=True,
    progress=False
)["Close"]


def compute_variance(prices, freq):

    resampled = prices.resample(freq).last()
    returns = np.log(resampled / resampled.shift(1)).dropna()

    variance = returns.var()
    volatility = returns.std()

    return variance, volatility


# -------------------------
# DAILY
# -------------------------

daily_returns = np.log(prices / prices.shift(1)).dropna()

daily_variance = daily_returns.var()
daily_vol = daily_returns.std()

# -------------------------
# OTHER FREQUENCIES
# -------------------------

weekly_variance, weekly_vol = compute_variance(prices, "W")
monthly_variance, monthly_vol = compute_variance(prices, "ME")
annual_variance, annual_vol = compute_variance(prices, "YE")

# -------------------------
# ANNUALIZE
# -------------------------

daily_variance *= 252
weekly_variance *= 52
monthly_variance *= 12
annual_variance *= 1

daily_vol *= np.sqrt(252)
weekly_vol *= np.sqrt(52)
monthly_vol *= np.sqrt(12)
annual_vol *= np.sqrt(1)

# -------------------------
# OUTPUT
# -------------------------

print("\nAnnualized Variance / Volatility (3-year sample)\n")

print("Daily-based:", daily_variance, daily_vol)
print("Weekly-based:", weekly_variance, weekly_vol)
print("Monthly-based:", monthly_variance, monthly_vol)
print("Annual-based:", annual_variance, annual_vol)
```
